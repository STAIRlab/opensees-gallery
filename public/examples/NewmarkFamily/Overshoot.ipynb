```python
import sdof
import numpy as np
import matplotlib.pyplot as plt
```

```python
period = 0.1
dt = 10*period
zeros = np.zeros(int(40//dt))
```

```python
u, v, a = sdof.sdof(period=period).integrate(zeros, dt, beta=0.3025, gamma=0.6, u0=1)
```

```python
plt.plot(u)
plt.plot(v)
```
